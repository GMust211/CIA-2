```python
import pandas as pd
column_name = 'car_name'
unique_values = df[column_name].unique()
mapping = {value: float(index + 1) for index, value in enumerate(unique_values)}
df[column_name] = df[column_name].map(mapping)
df
```

```python
# Fit the KMeans model
kmeans = KMeans(n_clusters=5)
label = kmeans.fit_predict(df)

# Get the centroids
centroids = kmeans.cluster_centers_

# Filter rows of original data
filtered_labels = [df[label == i] for i in range(5)]

# Define colors for each label
colors = ['red', 'blue', 'green', 'orange', 'purple']

# Plotting the results with specified colors and centroids
for i, (filtered_label, color) in enumerate(zip(filtered_labels, colors)):
    plt.scatter(filtered_label.iloc[:, 0], filtered_label.iloc[:, 1], color=color, label=f'Label {i}', s=8)

# Plot centroids
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', color='black', label='Centroids', s=30)

plt.legend()
plt.show()

grouped_data = df.groupby(label)

cluster_stats = []
for i, group in grouped_data:
    stats = group.describe()
    stats.index = pd.Index([f'Label {i}'] * len(stats))  # Add label to index for identification
    cluster_stats.append(stats)

# Concatenate statistics for all cluster groups
cluster_stats_df = pd.concat(cluster_stats)
cluster_stats_df
```
